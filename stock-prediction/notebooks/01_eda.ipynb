# Exploratory Data Analysis (EDA)
Explore raw market data, visualize price trends, and check for missing values.

```python
# Load and inspect raw data
ticker = 'AAPL'
from src.data_loader import download_data, clean_data
df = download_data(ticker, start='2015-01-01')
df = clean_data(df)
df.head()
```

```python
# Visualize closing price
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
df['Close'].plot(title=f'{ticker} Closing Price')
plt.show()
```

```python
# Check for missing values
df.isnull().sum()
```

## Assumptions
- Data is downloaded from yfinance and cleaned for missing values.
- No lookahead bias: all analysis uses only available historical data.
